```python

c=0
while c==0:
    x,y=map(int,input("Enter two numbers:").split(','))
    operation=input("Enter the operation to be performed:").upper()
    match operation:
        case "ADD":
            c=0
            print("Sum=",x+y)
            
        case "SUB":
            c=0
            print("Sub=",x-y)
        case "PROD":
            c=0
            print("Prod=",x*y)
        case "DIV":
            c=0
            print("Div=",round(x/y,2))
        case "MOD":
            c=0
            print("Mod=",x%y)
        case _:
            c=1
            print("Invalid Input")
    
```
